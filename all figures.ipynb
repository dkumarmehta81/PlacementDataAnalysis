```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Placement_Data_Full_Class_1.csv')
df.head()
```

```python
df.drop(['Unnamed: 0','sl_no'],axis=1,inplace=True)
```

```python
fig,axes=plt.subplots(3,2,figsize=(15,15),sharey=True)
plt.subplots_adjust(bottom=0, top=0.8, hspace=0.1)

sns.countplot(df.gender,hue=df.status,ax=axes[0,0])
axes[0,0].set_title("(a) Placement According to Gender",y=-0.18,fontweight='bold',y=-0.18,fontweight='bold')
axes[0,0].set_xlabel('Gender',fontsize=13)
axes[0,0].set_ylabel('Count',fontsize=13)
axes[0,0].tick_params(labelsize=12)

sns.countplot("ssc_b", hue="status", data=df,ax=axes[0,1])
axes[0,1].set_title("(b) Placement According to Senior Secondary Board",y=-0.18,fontweight='bold',y=-0.18,fontweight='bold')
axes[0,1].set_xlabel('SSC_B',fontsize=13)
axes[0,1].set_ylabel('Count',fontsize=13)
axes[0,1].tick_params(labelsize=12)

sns.countplot("hsc_b", hue="status", data=df,ax=axes[1,0])
axes[1,0].set_title("(c) Placement According to Higher Secondary Board",y=-0.18,fontweight='bold',y=-0.18,fontweight='bold')
axes[1,0].set_xlabel('HSC_B',fontsize=13)
axes[1,0].set_ylabel('Count',fontsize=13)
axes[1,0].tick_params(labelsize=12)

sns.countplot("hsc_s", hue="status", data=df,ax=axes[1,1])
axes[1,1].set_title("(d) Placement According to Higher Secondary Subject Stream",y=-0.18,fontweight='bold',y=-0.18,fontweight='bold')
axes[1,1].set_xlabel('HSC_S',fontsize=13)
axes[1,1].set_ylabel('Count',fontsize=13)
axes[1,1].tick_params(labelsize=12)

sns.countplot("degree_t", hue="status", data=df,ax=axes[2,0])
axes[2,0].set_title("(e) Placement According to Bechelor Degree Stream",y=-0.18,fontweight='bold',y=-0.18,fontweight='bold')
axes[2,0].set_xlabel('Degree_T',fontsize=13)
axes[2,0].set_ylabel('Count',fontsize=13)
axes[2,0].tick_params(labelsize=12)

sns.countplot("workex", hue="status", data=df,ax=axes[2,1])
axes[2,1].set_title("(f) Placement According to Future Work Experience",y=-0.18,fontweight='bold',y=-0.18,fontweight='bold')
axes[2,1].set_xlabel('WorkEx',fontsize=13)
axes[2,1].set_ylabel('Count',fontsize=13)
axes[2,1].tick_params(labelsize=12)
#axes.set_rasterize(True)
plt.tight_layout()
plt.savefig("fig4.eps")
```

```python
from imblearn.over_sampling import SMOTE
X=df.drop(['gender'],axis=1)
y=df['gender']
```

```python
X.status=pd.get_dummies(X.status,drop_first=True)
```

```python
X.ssc_b=pd.get_dummies(X.ssc_b,drop_first=True)
X.hsc_b=pd.get_dummies(X.hsc_b,drop_first=True)
X.hsc_s=pd.get_dummies(X.hsc_s,drop_first=True)
X.workex=pd.get_dummies(X.workex,drop_first=True)
X.specialisation=pd.get_dummies(X.specialisation,drop_first=True)
degree_t=pd.get_dummies(df.degree_t,drop_first=True)
X=pd.concat([X,degree_t],axis=1)
X.drop("degree_t",axis=1,inplace=True)
```

```python
y=y.map({'M':0,'F':1})
```

```python
sm=SMOTE(random_state=0)
X,y=sm.fit_sample(X,y)
```

```python
fig,axes=plt.subplots(1,2,figsize=(12,5),sharey=True)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
labels=['M','F']
label_index=[0,1]
df.gender.value_counts().plot(kind="bar",color=colors,ax=axes[0])
axes[0].set_title('(a) Placement Gender Class',fontsize=15,y=-0.18,fontweight='bold')
axes[0].set_xlabel('Gender',fontsize=13)
axes[0].set_ylabel('Frequency',fontsize=13)
axes[0].tick_params(labelsize=12)
#axes[0,0].set_xticks(label_index,labels,fontsize=12)

y.value_counts().plot(kind="bar",color=colors,ax=axes[1])
axes[1].set_title('(b)Placement Gender Class After Re-Sampling',fontsize=15,y=-0.18,fontweight='bold')
axes[1].set_xlabel('Gender',fontsize=13)
axes[1].set_ylabel('Frequency',fontsize=13)
axes[1].tick_params(labelsize=12)
#axes[1].xticks(label_index,labels,fontsize=15)
axes[1].set_xticklabels(labels, fontdict=None, minor=False)

plt.tight_layout()
plt.savefig('fig8.eps')
```
